```python
import pandas as pd

# Load dataset
df = pd.read_csv("used_cars_messy.csv")

# Basic exploration
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nDescribe:")
print(df.describe(include='all'))
```

```python
df = df.dropna(subset=['selling_price'])

# Clean 'km_driven' column by removing non-numeric characters and converting to numeric
df['km_driven'] = df['km_driven'].str.replace(',', '', regex=False) # Remove commas if any
df['km_driven'] = pd.to_numeric(df['km_driven'], errors='coerce') # Convert to numeric, coercing errors to NaN
df['km_driven'] = df['km_driven'].fillna(df['km_driven'].median())

df['mileage'] = df['mileage'].fillna(df['mileage'].mode()[0])

df['brand'] = df['brand'].str.strip().str.lower()

df['mileage'] = df['mileage'].str.extract('(\\d+\\.?\\d*)').astype(float)

df = df.drop_duplicates()

print("Cleaned Shape:", df.shape)
```

```python
from sklearn.metrics import mean_absolute_error

mean_price = df['selling_price'].mean()

y_true = df['selling_price']
y_pred = [mean_price] * len(df)

mae = mean_absolute_error(y_true, y_pred)

print("Baseline MAE:", mae)
```
